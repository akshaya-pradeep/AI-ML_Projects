```python
def removal_box_plot(df,column,threshold):
  sns.boxplot(df[column])
  plt.title(f"Original box plot of {column}")
  plt.show()

  removed_outliers=df[df[column] >= threshold and df[column] <= threshold]

  sns.boxplot(removed_outliers[column])
  plt.title(f"Box plot without outliers of {column}")
  plt.show()

  return removed_outliers

threshold_value=[45,45,10,16,20,60,10,30,50,12,50,1,2000,0.1,15,16,0,18,3]

no_outliers = removal_box_plot(df,column,threshold_value[i])
```
