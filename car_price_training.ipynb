```python
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],dtype=int,drop_first=True)
df1
```

We have to concatinate the old df and encoded new df(df1)

```python
new_df=pd.concat([df,df1],axis=1)
new_df
```

```python
# drop all the unwanted columns
# Location,Brand,Transmission etc are dropping bcz that are already encoded and in df1 so we combined the results so no need of that columns again
new_df=new_df.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','Brand','New_Price'],axis=1)
new_df
```

```python
# removing the units, after that we will convert the object dtype to numerical
new_df['Mileage']=new_df['Mileage'].str.replace('km/kg','') # replacing with a space
new_df['Mileage']=new_df['Mileage'].str.replace('kmpl','')
new_df['Engine']=new_df['Engine'].str.replace('CC','')
new_df['Power']=new_df['Power'].str.replace('bhp','')
```

```python
new_df.head()
```

```python
new_df.dtypes
```

```python
# there are typed null (its mentioned in q), they are not actually missing values
# we have to replace the with 0
new_df['Mileage']=new_df['Mileage'].str.replace('null','0')
new_df['Engine']=new_df['Engine'].str.replace('null','0')
new_df['Power']=new_df['Power'].str.replace('null','0')
new_df
```

```python
new_df.dtypes
```

When removing the units it wont be converted to numerical

So Change the dtype of mileage,engine and power

```python
new_df['Mileage']=new_df['Mileage'].astype(float)
new_df['Engine']=new_df['Engine'].astype(float)
new_df['Power']=new_df['Power'].astype(float)
```

```python
new_df.dtypes
```

Mileage,Engine,Power of a vehicle cannot be 0.

So we have to replace the values as missing values.

cannot replace a string as missing value thats y converted before as numerical to zero


```python
new_df.loc[new_df.Engine==0,'Engine']=np.NaN
new_df.loc[new_df.Power==0,'Power']=np.NaN
new_df.loc[new_df.Mileage==0,'Mileage']=np.NaN
```
