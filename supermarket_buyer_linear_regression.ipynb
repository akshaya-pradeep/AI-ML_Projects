```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y,color='g')
plt.xlabel('Visitor')
plt.ylabel('Buyer')
plt.title('Visitor-Buyer Graph')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_Value':y_test,'Predicted_Value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
plt.scatter(x_train,y_train,color='c')
plt.plot(y_test,y_pred,color='k')
plt.xlabel('Visitors visited the shop')
plt.ylabel('Out of the visitor how many actually buy')
plt.title('Visitors vs Buyers Graph')
```
