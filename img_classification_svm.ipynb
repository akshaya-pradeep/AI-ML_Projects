```python
flat_data_arr = []           # To collect input samples
target_arr = []              # To collect output samples
categories = ['Cat','Dog']   #
datadir = '/content/drive/MyDrive/Data-20241230T070359Z-001/Data'
```

```python
for i in categories:       # first cat then dog
  print("Loading categories..........") # just to know it is loading we print a msg
  path = os.path.join(datadir,i)  # /content/drive/MyDrive/Data-20241230T070359Z-001/Data,Cat
  for img in os.listdir(path):
    img_arr = imread(os.path.join(path,img))
    img_resize = resize(img_arr,(150,150,3)) # std size, 150x150x3 columns, 150->height, 150->width, 3->RGB color format
    flat_data_arr.append(img_resize.flatten()) # to convert to linear(1D)
    target_arr.append(categories.index(i)) # cat - 0 , dog - 1 , indexes
  print("Loaded completely",i)
```

```python
flat_data = np.array(flat_data_arr) # converting into array
target_data = np.array(target_arr)

df = pd.DataFrame(flat_data)
df.shape
```

```python
df['Target'] = target_data
df.shape
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
acc_score = accuracy_score(y_test,y_pred)
acc_score
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
