```python
train['Age']=train['Age'].fillna(train['Age'].mean())
train['Cabin']=train['Cabin'].fillna(train['Cabin'].mode()[0])
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
train.isna().sum()
```

```python
train.dtypes
```

```python
for col in train.dtypes[train.dtypes == 'object'].index:
  for_dummy=train.pop(col)
  train=pd.concat([train,pd.get_dummies(for_dummy,prefix=col)],axis=1)
train.head()
```

```python
x=train.drop(['Survived'],axis=1)
x
```

```python
y=train['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_curve,roc_auc_score #Reciever Operating Characteristic Curve
cm=confusion_matrix(y_test,y_pred)
acc_score=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
false_positive_rate,true_positive_rate,thresholds=roc_curve(y_test,y_pred)
roc_auc_curve=roc_auc_score(y_test,y_pred)
```

The ROC curve visualizes the trade-off between the True Positive Rate (TPR) and False Positive Rate (FPR) as the decision threshold varies.
The AUC (Area Under the Curve) tells us how well the model is distinguishing between the two classes.



The diagonal line (baseline) represents a model that makes random predictions.
A perfect model will have an ROC curve that moves far away from this diagonal, achieving a larger area under the curve (AUC).
If the ROC curve is close to the diagonal, the model is not doing better than random guessing.

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(false_positive_rate, true_positive_rate, color='r', label=f'ROC Curve (AUC = {roc_auc_curve:.4f})') # Adds a label to the curve, including the AUC value formatted to 4 decimal places.

# Plot the random guessing line (diagonal)
plt.plot([0, 1], [0, 1], color='k', linestyle='--', label='Random Guessing') #Plots a diagonal line from (0, 0) to (1, 1).

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
