```python
import numpy as np
import matplotlib.pyplot as plt
area=[2400,2600,3000,3400,4000]
price=[5500000,5850000,6200000,6700000,7200000]
plt.scatter(area,price,color='m')

plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area-Price Graph of a House')
```

After marking all the points, we have to draw a straight line.

- We cannot draw a straight line passing through all the points
- so machine will draw a straight line which approximately passes through majority of points
- **That line is called Regression line**
- Based on the regression line the algorithm works

→ First we have to draw a perpendicular line from x axis(area is in x axis and for some area we have to predict the price) that will touch the straight line.

→ Then draw a line parallel to x axis and perpendicular to y axis.

→ That value will be the predicted price of that area.

### Errors

- in regression model we calculate errors not accuray or anything
- Errors are calculated by taking the difference between actual value and predicted value
- Errors need to be optimised
