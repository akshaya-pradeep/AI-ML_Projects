# Box Plot

- To find/identify outliers (wrong datas) box plot is used.

```python
sns.boxplot(x='smoker',y='total_bill',data=df,color='m')
```

circles are the outliers.

Rest in note

Suppose we have datas 3,5,7,1,1,8,4,6

â†’ First sort the datas

1,1,3,4,      5,6,7,8

â†’ Calculate the median

Median = (4+5)/2 = 4.5 â‡’ Q2 = 4.5

â†’ Calculate the median of 1st part

Median = (1+3)/2 = 2 â‡’ Q1 =2

â†’ Calculate the median of 2nd part

Median = (6+7)/2 = 6.5 â‡’ Q3

â†’ Calculate **IQR (Inter Quartile Range)**

IQR = Q3 - Q1  â‡’  6.5 - 2 = 4.5

â€”> Outliers can be seen in top and botto,

â†’ Calculate Top Outliers = (1.5xIQR+Q3)  = 13.25

- Ie if a value comes above 13.25 it will be considered as outliers.

â†’ Calculate Bottom Outliers = (Q1-.1.5xIQR)  =  -4.75

- Values that comes below -4.75 are considered as bottom outliers.

```python
sns.boxplot(y='total_bill',data=df)
```

```python
import matplotlib.pyplot as plt
def removal_box_plot(df,column,threshold):
  sns.boxplot(df[column])
  plt.title(f"Original box plot of {column}")
  plt.show()

  removed_outliers=df[df[column] >= threshold]

  sns.boxplot(removed_outliers[column])
  plt.title(f'Box Plot Without Outliers of {column}')
  plt.show()
  return removed_outliers

threshold_value=40

no_outliers = removal_box_plot(df,'total_bill',threshold_value)
```

Without plotting box plot we can find outliers as

```python
import numpy as np
Q1 = np.percentile(df['total_bill'],25,method='midpoint') #25 Q1
Q3 = np.percentile(df['total_bill'],75,method='midpoint') # 75 Quarter3
IQR = Q3 - Q1
print(IQR)
```

```python
upper = Q3 + 1.5 * IQR
upper_array = np.array(df['total_bill'] >= upper)
print("Upper Bound = ",upper)
print("No.of outliers = ",upper_array.sum())

# BElow lower bound
lower = Q1 - 1.5 * IQR
lower_array = np.array(df['total_bill'] <= lower)
print("lower Bound = ",lower)
print("No.of outliers = ",lower_array.sum())
```
