SUB PLOT

```python
# subplot
# There will be two partitions
# partition1 used for graph paper1, partition2 used for graph paper2
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1) # 1st num:no.of rows, 2nd num:no.of columns, 3rd num: position of graph(1 or 2)
plt.plot(x1,y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph-1')

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph-2')
```

```python
# To add main tittle
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph-1')

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph-2')

plt.suptitle('Graph_1-Graph_2')
```

```python
y=[10,20,30,40]
label=['apple','orange','mango','grape']
plt.pie(y,labels=label)
plt.legend(loc='lower left') # upper right,left , lower left, right
```

```python
y=[10,20,30,40]
label=['apple','orange','mango','grape']
plt.pie(y,labels=label)
plt.legend(loc='lower left')
plt.tight_layout() #to avoid overlapping
```

```python
y=[10,20,30,40]
label=['apple','orange','mango','grape']
plt.pie(y,labels=label,autopct='%1.01f%%') # 1->integer value and 01-> decimal 2 valus,  % symbol
plt.legend(loc='lower left')
plt.tight_layout()
```

```python
y=[10,20,30,40]
label=['apple','orange','mango','grape']
data=[0.2,0,0,0.3] # deviation
plt.pie(y,labels=label,autopct='%1.01f%%',explode=data)
plt.legend(loc='lower left')
plt.tight_layout()
```
