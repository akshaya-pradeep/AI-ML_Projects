```python
# dtype of trestbps is float, but we cant simply take mean. First check the unique values
df['trestbps'].unique()
# Lot of values so can use mean
```

```python
# trest_mean=df['trestbps'].mean()
# df['trestbps'].fillna(trest_mean,inplace=True)
df['trestbps']=df['trestbps'].fillna(df['trestbps'].mean())
```

```python
df['restecg'].unique()
# Only some values so we have to use mode for filling missing values
```

```python
df['restecg']=df['restecg'].fillna(df['restecg'].mode()[0])
```

```python
df['thalach'].unique()
```

```python
df['thalach']=df['thalach'].fillna(df['thalach'].mean())
```

```python
df['ca'].unique()
```

```python
df['ca']=df['ca'].fillna(df['ca'].mode()[0])
```

```python
df['thal'].unique()
```

```python
df['thal']=df['thal'].fillna(df['thal'].mode()[0])
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
acc_score=accuracy_score(y_test,y_pred)
cm
```

```python
acc_score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
