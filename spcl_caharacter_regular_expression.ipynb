```python
import re
str1='Wonderful @Peacock#$12345$^&* luminar@123!!!'
str2=re.sub('[a-z]',' ',str1)
str2
```

```python
str2=re.sub('[A-Z]',' ',str1)  # [a-z]  [A-Z] -> Patterns in regular expression
str2
```

```python
str2=re.sub('[0-9]',' ',str1)
str2
```

```python
str2=re.sub('[A-Z0-9a-z]',' ',str1) # it will contain only special characters
str2
```

```python
str2=re.sub('[^A-Z0-9a-z]',' ',str1)   # it will collect only special charaters and that ^ will remove those special characters
str2
```
