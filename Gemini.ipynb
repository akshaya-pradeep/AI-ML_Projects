```python
import google.generativeai as genai
```

```python
from google.colab import userdata
google_api_key = userdata.get('GOOGLE_API_KEY')
```

```python
# configuration
genai.configure(api_key=google_api_key)
```

```python
# To list the models in gemini
for model in genai.list_models():
  # print(model)
  print(model.name)
```

```python
# we just use models/gemini-pro now
model = genai.GenerativeModel('gemini-1.5-pro')
```

```python
response = model.generate_content("What is machine learning")
print(response.text) # To get the output as text, it will remove the extra json contents etc
```

```python
from IPython.display import Markdown
import textwrap  # bold etc is defined in textwrap

def demo(text):
  return Markdown(textwrap.indent(text,'>')) # > means go to next line

demo(response.text)
```

```python
response1 = model.generate_content("Luminar Technolab")
demo(response1.text)
```
