```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/lung_cancer_examples.csv')
df
```

```python
df.drop('Name',axis=1,inplace=True)
df.drop('Surname',axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_train.shape
```

```python
x_test
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
acc_score=accuracy_score(y_test,y_pred)
cm
```

```python
acc_score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
smokes_result = df.groupby(['Smokes', 'Result']).size().unstack()

smokes_result.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='coolwarm'
)
plt.title('Smoking Levels by Cancer Result')
plt.xlabel('Smoking Level')
plt.ylabel('Count')
plt.legend(title='Cancer Result', labels=['No', 'Yes'])
plt.show()
```

```python
import seaborn as sns

alkhol_mean = df.groupby('Result')['Alkhol'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=alkhol_mean.index, y=alkhol_mean.values, palette="muted")
plt.title('Average Alcohol Consumption vs. Cancer Result')
plt.xlabel('Cancer Result (0 = No, 1 = Yes)')
plt.ylabel('Average Alcohol Consumption')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
