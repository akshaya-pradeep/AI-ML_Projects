```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y,color='m')
plt.xlabel('Hours spent for study')
plt.ylabel('Score')
plt.title('Student hour-score graph')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actaul_Value':y_test,'Predicted_Value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print(model.predict([[9.3]]))
```

```python
plt.scatter(x_train,y_train,color='g')
plt.plot(x_test,y_pred,color='r')
plt.xlabel('Hours studied')
plt.ylabel('Score Obtained')
plt.title('Hour spent - Score Obtained')
```

```python
print('Slope = ',model.coef_)
```

```python
print('y-intercept = ',model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
print('MAE = ',mean_absolute_error(y_test,y_pred))
```

```python
print('MAPE = ',mean_absolute_percentage_error(y_test,y_pred))
```

```python
print('MSE = ',mean_squared_error(y_test,y_pred))
```

```python
print('RMSE = ',root_mean_squared_error(y_test,y_pred))
```

```python
print('R2_Score = ',r2_score(y_test,y_pred))
```
