```python
from diffusers import StableDiffusionPipeline,EulerAncestralDiscreteScheduler
```

```python
model_id = 'stabilityai/stable-diffusion-2'
```

```python
# load model
scheduler = EulerAncestralDiscreteScheduler.from_pretrained(model_id,subfolder='scheduler')
pipe = StableDiffusionPipeline.from_pretrained(model_id,scheduler=scheduler) # this scheduler is the variable name above
```

```python
pipe = pipe.to('cuda')
```

```python
def txtimg(prompt):
  img = pipe(prompt,guidance_scale=9,height=768,width=768).images[0]
  return img
```

```python
prompt = 'A dog and cat wearing helmet in sky'
txtimg(prompt)
```

```python
import gradio as gr
demo = gr.Interface(fn=txtimg,inputs='text',outputs='image')
demo.launch()
```
