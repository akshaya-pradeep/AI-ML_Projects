```python
from sklearn.compose import make_column_transformer # to expand each calumn label as independent columns
from sklearn.preprocessing import OneHotEncoder
end=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=end.fit_transform(x)
x

# handle_unknown='ignore'  --> To ignore unknown states
# remainder='passthrough'  --> To pass the rest of the columns
# the column needs to be encoded is state
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print(list(zip(x,model.coef_)))
```

```python
print('y intercept = ',model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
print('MAE = ',mean_absolute_error(y_test,y_pred))
print('MAPE = ',mean_absolute_percentage_error(y_test,y_pred))
print('MSE = ',mean_squared_error(y_test,y_pred))
print('RMSE = ',root_mean_squared_error(y_test,y_pred))
print('R2_score = ',r2_score(y_test,y_pred))
```
