5.Read face cream and facewash product sales data and show it using the bar chart with grid line

```python
# -,+ is used to arrange at left and right respectively
plt.bar([a-0.25 for a in month], face_cream, width=0.25, label='Face Cream Sales Data', align='edge') # align -> to arrange at edge without overlapping
plt.bar([a+0.25 for a in month], face_wash, width=-0.25, label='Face Wash Sales Data', align='edge')
plt.xlabel('Month Numer')
plt.ylabel('Sales units in number')
plt.legend(loc='upper right')
plt.title('Sales Data')
plt.xticks(month)
plt.title('Face wash and Face Cream Sales Data')
# plt.show()
```

```python
# barchart b/w toothpaste and bathingsoap
plt.bar([a-0.25 for a in month], tooth_paste, width= 0.25, label='Tooth Paste Sales Data', align='edge')
plt.bar([a+0.25 for a in month], bathing_soap, width= -0.25, label='Bathing Soap Sales Data', align='edge')
plt.ylabel('Sales units in number')
plt.legend(loc='upper right')
plt.title('Sales Data')
plt.xticks(month)
plt.title('Tooth Paste and Bathing Soap Sales Data')
```

7.Calculate total sale data for last year for each product and show it using a Pie chart

```python
lab=['face_cream','face_wash','tooth_paste','bathing_soap','shampoo','moisturizer']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
my_color=['r','g','b','m','c','y']
plt.pie(sales,labels=lab,autopct='%1.01f%%',colors=my_color)
plt.legend(loc='upper right')
plt.title('Sales Data')
plt.tight_layout()
```
