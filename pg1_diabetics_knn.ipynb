## Normalization or scaling

Sometimes the machine may ignore smaller values when there is a higher value. in this eg glucose level is high and predegree is low. so machine may remove that.

but that will lead to wrong prediction. and each input in that case is very important. so we have to normalise the value so the machine may not differentiate the inputs
based on the range.

 1. Standard scalar
  z = (x-u)/s

z => output after normalisation

x => each input sample

u => mean of training data

s => std deviation of training data

in case of glucose z=(148-mean of training data of glucose)/std deviation of traing data of glucose

 2. Minmax scalar

 these two are based on mathematical formulae


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model creation (knn model) default k value = 5
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred  # predicted output, actual o/p will be in y_test
```
